```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file1 = 'theater_complaints.csv'
file2 = 'Customer-Gender.csv'
file3 = 'student_scores_data.csv'
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)
```

```python
print("Dataset preview:")
print(df1.head())

print("\nDataset summary:")
print(df2.describe())
```

```python
# Average number of complaints registered
average = df1['Number of complaints'].mean()
print(f"\nAverage number of complaints registered : ",int(average))
```

```python
ns=df2[df2['CustomerGender']=='NotSpecified']
print(ns)
```

```python
EntGen=df2['CustomerGender'].replace('NotSpecified','Enter Gender')
print(EntGen)
```

```python
# 1. Bar chart of Theator complaints
df1.plot(kind='bar', title='Theator complaints',x='Movie Theater Goers Complaints',y='Number of complaints',color='skyblue')
plt.xlabel('Movie Theater Goers Complaints')
plt.ylabel('Number of complaints')
plt.show()
```

```python
# 2. Scatter plot between 'SAT' & 'GPA'
x_col, y_col = 'SAT', 'GPA'
df3.plot.scatter(x=x_col, y=y_col, title=f'Scatter Plot of {x_col} vs {y_col}', color='orange')
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.show()
```

```python
# 3. Heatmap of correlations between numerical columns
cols=['Number of complaints' , 'frequency']
corr=df1[cols]
plt.figure(figsize=(10, 8))
plt.imshow(corr, cmap='Greens');
plt.colorbar();
plt.title('Heatmap');
plt.show()
```

```python
print("\nObservations and insights:")
print("- The bar chart shows the distribution of values of Number of complaints registered by Theator Goers.")
print("- The scatter plot highlights potential relationships between Students' scores.")
print("- The heatmap gives the relatioship between Number of complaints registered & Frequency of occurance.")

```

```python
import numpy as np


def get_matrix_input(prompt):
    print(prompt)
    rows = int(input("Enter the number of rows: "))
    cols = int(input("Enter the number of columns: "))
    print(f"Enter the elements of the matrix (row by row):")
    elements = []
    for i in range(rows):
        row = list(map(float, input(f"Row {i+1}: ").split()))
        if len(row) != cols:
            print("Error: Number of elements does not match column count. Try again.")
            return get_matrix_input(prompt)
        elements.append(row)
    return np.array(elements)

def display_menu():
    print("\nMatrix Operations Tool")
    print("1. Add Matrices")
    print("2. Subtract Matrices")
    print("3. Multiply Matrices")
    print("4. Transpose Matrix")
    print("5. Determinant of a Matrix")
    print("6. Exit")

def main():
    while True:
        display_menu()
        choice = input("\nSelect an operation (1-6): ")

        if choice == "1":
            print("\nAddition of Matrices")
            matrix_a = get_matrix_input("Enter the first matrix:")
            matrix_b = get_matrix_input("Enter the second matrix:")
            if matrix_a.shape == matrix_b.shape:
                result = matrix_a + matrix_b
                print("\nResult:")
                print(result)
            else:
                print("Error: Matrices must have the same dimensions.")

        elif choice == "2":
            print("\nSubtraction of Matrices")
            matrix_a = get_matrix_input("Enter the first matrix:")
            matrix_b = get_matrix_input("Enter the second matrix:")
            if matrix_a.shape == matrix_b.shape:
                result = matrix_a - matrix_b
                print("\nResult:")
                print(result)
            else:
                print("Error: Matrices must have the same dimensions.")

        elif choice == "3":
            print("\nMultiplication of Matrices")
            matrix_a = get_matrix_input("Enter the first matrix:")
            matrix_b = get_matrix_input("Enter the second matrix:")
            if matrix_a.shape[1] == matrix_b.shape[0]:
                result = np.dot(matrix_a, matrix_b)
                print("\nResult:")
                print(result)
            else:
                print("Error: Number of columns of the first matrix must equal the number of rows of the second matrix.")

        elif choice == "4":
            print("\nTranspose of a Matrix")
            matrix = get_matrix_input("Enter the matrix:")
            result = matrix.T
            print("\nResult:")
            print(result)

        elif choice == "5":
            print("\nDeterminant of a Matrix")
            matrix = get_matrix_input("Enter the matrix (must be square):")
            if matrix.shape[0] == matrix.shape[1]:
                result = np.linalg.det(matrix)
                print("\nResult:")
                print(result)
            else:
                print("Error: Matrix must be square.")

        elif choice == "6":
            print("Exiting the Matrix Operations Tool. Goodbye!")
            break

        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
